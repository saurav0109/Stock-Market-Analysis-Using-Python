```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
#aapl = yf.Ticker("AAPL")
# AMZN,GOOGlE,META
```

```python
from datetime import datetime
end = datetime.now()
```

```python
#stocksymbols = ["AAPL","GOOG","MSFT","AMZN"]
```

```python
start = datetime(end.year-2,end.month,end.day-1)
end_d = datetime(end.year,end.month,end.day-1)
```

```python
tick1='AMZN'
tick2='AAPL'
tick3='GOOG'
tick4='MSFT'
```

```python
AMZN=yf.download(tick1,start,end_d)
AAPL=yf.download(tick2,start,end_d)
GOOG=yf.download(tick3,start,end_d)
MSFT=yf.download(tick4,start,end_d)
```

# APPLE 

```python
AAPL.info()
```

```python
AAPL.head(10)
```

```python
AAPL['Close'].plot(figsize=(15,5))
plt.ylabel('Price($)')
plt.xlabel('Year')
plt.show()
```

```python
AAPL['Volume'].plot(figsize=(15,5))
plt.ylabel('Price($)')
plt.xlabel('Year')
plt.show()
```

```python
window_size = [7,15,30]

for ws in window_size :
    col_name = f"Ma for{ws}day"
    AAPL[col_name]=AAPL['Close'].rolling (window = ws, center = False).mean()
```

```python
AAPL.head(35)
```

```python
AAPL['Daily Return'] = AAPL['Close'].pct_change()
```

```python
AAPL.head(20)
```

```python
AAPL['Trend'] = pd.cut(AAPL['Daily Return'], bins=[float('-inf'), 0, float('inf')], labels=['Down', 'Up',])
```

```python
print(AAPL)
```

```python
category_counts = AAPL['Trend'].value_counts()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99'])
plt.title('Trend Frequency Pie Chart')
plt.show()
```
